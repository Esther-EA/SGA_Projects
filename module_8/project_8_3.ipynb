**Check for missing values and do what you feel is appropriate to them**

```python
df.isnull()
```

```python
df['RATING'].fillna(df['RATING'].mean(), inplace=True)
df.head()
```

```python
df["RunTime"].fillna(df["RunTime"].mean(), inplace = True)
df.head(10)
```
