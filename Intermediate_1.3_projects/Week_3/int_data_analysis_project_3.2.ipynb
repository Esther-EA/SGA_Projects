Data Cleaning
Replace all the column names that has space(s) between them with an Underscore

```python
# remove spaces from column names
# give col a var
df_col = df2.columns
# loop over the column names
new_col = []
for col in df_col:
  if " " in col:
    x= col.replace(" ", "_")
    new_col.append(x)
  else:
    new_col.append(col)
print(new_col)
```

```python
df2.columns = new_col
df2.head()
```

# Cleaning Columns and Change column data types
Some columns data are integer but were represented as an object on the DataFrame

```python
cleaned_Air = df2.Air_Pollution_Score.apply(lambda x: x.split("/")[0])
```

```python
cleaned_Air =cleaned_Air.astype(float)
```

```python
df2.Air_Pollution_Score = cleaned_Air
```

```python
df2.head()
```

```python
# define a function the can be reused for cleaning other columns
def convert_to_number(column_name):
    total_string = []
    for i in column_name:
        if "/" in i:
            new_string = int(i.split("/")[1])
            total_string.append(new_string)
        else:
            total_string.append(int(i))
    return total_string
```

```python
# Apply function to other columns
City_MPG = convert_to_number(df2.City_MPG)
df2.City_MPG = City_MPG
```

```python
Hwy_MPG = convert_to_number(df2.Hwy_MPG)
df2.Hwy_MPG = Hwy_MPG
```

```python
Greenhouse_Gas = convert_to_number(df2.Greenhouse_Gas_Score)
df2.Greenhouse_Gas_Score = Greenhouse_Gas
```

```python
# confirm that the data types has been changed
df2.dtypes
```
